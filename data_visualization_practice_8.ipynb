```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset
header=['sepal_lenght', 'sepal_width', 'petal_lenght', 'petal_width', 'specie']

tip = pd.read_csv('/content/drive/MyDrive/Datasets/exam 1/a/tips.csv')
iris = pd.read_csv('/content/drive/MyDrive/Datasets/pandas 3/iris.data', header=0)
iris.columns = header
flights = pd.read_csv('/content/drive/MyDrive/Datasets/pandas 4/flights.csv')
```

```python
# 1
sns.barplot(x='day', y='tip', data=tip, estimator='mean', color='#dbc81a')
plt.xlabel('Tip mean')
plt.ylabel('Day of Week')
plt.title('Tip mean by day of week')

plt.show()
```

```python
# 2
sns.scatterplot(x='tip', y='total_bill', data=tip, color='#dbc81a')
plt.title('Total bill by tip')

plt.show()
```

```python
# 3
sns.pairplot(iris, hue='specie', palette='Wistia')

plt.show()
```

```python
# 4
sns.scatterplot(x='tip', y='total_bill', data=tip, color='#dbc81a', alpha=0.5)
plt.title('Total bill by tip')

plt.show()
```

```python
# 5
sns.boxplot(x='time', y='tip', data=tip, color='yellow')
plt.title('Tip by time')

plt.show()
```

```python
# 6
corr = flights[['MONTH', 'ARRIVAL_DELAY', 'ELAPSED_TIME', 'AIR_TIME', 'DISTANCE']].corr()

sns.heatmap(corr, cmap='Wistia')

plt.show()
```

```python
# 7
step = 10000
sns.lineplot(x='MONTH', y='DISTANCE', data=flights, color='#dbc81a')
plt.title('Distance traveled by month')

plt.show()
```

```python
# 8
to_violin = tip.iloc[:]
to_violin
sns.violinplot(x='time', y='total_bill', data=to_violin, fill=False, hue='time', palette='Wistia')
plt.title('Total bill by time')

plt.show()
```

```python
# 9
g = sns.FacetGrid(tip, col='time', row='sex')
g.map(sns.scatterplot, 'total_bill', 'tip', color='#dbc81a')

plt.suptitle('FaceGrid of scatterplot of tip by time and sex')
plt.tight_layout()
plt.show()
```

```python
# 10
sns.jointplot(x='sepal_lenght', y='sepal_width', data=iris, kind='kde', color='#dbc81a')

plt.suptitle('Jointplot of sepal lenght by sepal width with KDE kind')
plt.tight_layout()
plt.show()
```
