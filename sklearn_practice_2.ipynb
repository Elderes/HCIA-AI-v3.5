```python
# 1
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# Load data
iris = load_iris()
x = pd.DataFrame(iris.data, columns=iris.feature_names)
y = iris.target

# Apply PCA
pca = PCA(n_components=2)
x = pca.fit_transform(x)

# Plot
plt.figure()
colors = ['navy', 'turquoise', 'darkorange']
target_names = iris.target_names
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(x[y == i, 0], x[y == i, 1], color=color, alpha=0.8, label=target_name)
plt.legend(loc='best', scatterpoints=1)
plt.title('PCA of iris dataset')
plt.show()
```

```python
# 2
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# Load data
iris = load_iris()
x = pd.DataFrame(iris.data, columns=iris.feature_names)
y = iris.target

# Apply PCA
pca = PCA(n_components=2)
x = pca.fit_transform(x)
variance_ratios = pca.explained_variance_ratio_

# Plot
plt.figure()
plt.bar(range(len(variance_ratios)), variance_ratios, color='teal')
plt.xlabel('principal components')
plt.ylabel('variance explained')
plt.title('Variance explained by each principal component')
plt.show()
```

```python
# 3
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA

# Load data
wine = load_wine()
x = pd.DataFrame(wine.data, columns=wine.feature_names)
y = wine.target

# Apply PCA
pca = PCA(n_components=2)
z = pca.fit_transform(x)

# Plot
target_names = wine.target_names
colors = ['red', 'purple', 'darkred']
for colors, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(z[y == i, 0], z[y == i, 1], color=colors, label=target_name)
plt.legend(loc='best')
plt.show()
```

```python
# 4
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits

# Load data
digits = load_digits()
x = pd.DataFrame(digits.data, columns=digits.feature_names)
y = digits.target

# Apply t-SNE
tsne = TSNE(n_components=2)
x_tsne = tsne.fit_transform(x)

# Plot
colors = plt.cm.rainbow(np.linspace(0, 1, 10))
plt.figure()
for color, i in zip(colors, range(10)):
    plt.scatter(x_tsne[y == i, 0], x_tsne[y == i, 1], color=color, label=i)
plt.legend(loc='best')
plt.title('t-SNE visualization of the Digits dataset')
plt.show()
```

```python
# 5
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

# Load data
breast = load_breast_cancer()
x = breast.data
y = breast.target

# Apply PCA and LDA
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x)

lda = LDA(n_components=min(2, len(set(y)) - 1))
x_lda = lda.fit_transform(x, y)

# Plot
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].scatter(x_pca[:, 0], x_pca[:, 1], c=y, cmap='summer')
ax[0].set_title('PCA on Breast Cancer Dataset')
ax[1].hist(x_lda[y == True], color='yellow', alpha=0.7)
ax[1].hist(x_lda[y == False], color='green', alpha=0.5)
ax[1].set_title('LDA on Breast Cancer Dataset')
plt.show()
```

```python
# 6
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import KernelPCA

# Load data
iris = load_iris()
x = iris.data
y = iris.target

# Apply kernel PCA with RBF kernel
kpca = KernelPCA(n_components=2, kernel='rbf')
x_kpca = kpca.fit_transform(x)

# Plot
plt.scatter(x_kpca[:, 0], x_kpca[:, 1], c=y, cmap='rainbow')
plt.title('Kernel PCA with RBF kernel in Iris dataset')
plt.colorbar(label='specie')
plt.show()
```

```python
# 7
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import MDS

# Generate synthetic data
x_synthetic = np.random.rand(100, 10)

# Apply MDS
mds = MDS(n_components=2, normalized_stress='auto')
x_mds = mds.fit_transform(x_synthetic)

# Plot
plt.scatter(x_mds[:, 0], x_mds[:, 1], color='teal')
plt.title('MDS of synthetic data')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.show()
```

```python
# 8 
import matplotlib.pyplot as plt
from sklearn.manifold import Isomap
from sklearn.datasets import fetch_olivetti_faces

# Load data
faces = fetch_olivetti_faces()
x = faces.data
y = faces.target

# Apply Isomap
isomap = Isomap(n_components=2, n_neighbors=7)
x_isomap = isomap.fit_transform(x)

# Plot
for i in range(20):
    plt.scatter(x_isomap[i, 0], x_isomap[i, 1], marker='${}$'.format(i), s=100)
plt.title('Isomap in Olivetti Face Dataset')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.show()
```

```python
# 9
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_openml

# Load data
boston = fetch_openml('boston', version=1, parser='auto')
x = pd.DataFrame(boston.data, columns=boston.feature_names)
y = boston.target

# Apply PCA
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x)

# Get loadings
loadings = pca.components_.T * np.sqrt(pca.explained_variance_)

# Create a dataframe of loadings
loading_matrix = pd.DataFrame(loadings, columns=['PC1', 'PC2'], index=boston.feature_names)

# Plot
plt.bar(loading_matrix.index, loading_matrix['PC1'], label='PC1 Loadings')
plt.bar(loading_matrix.index, loading_matrix['PC2'], label='PC2 Loadings')
plt.title('PCA Feature Importance')
plt.ylabel('Loadings')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 10
import matplotlib.pyplot as plt
from sklearn.datasets import make_swiss_roll
from sklearn.manifold import LocallyLinearEmbedding

# Generate swiss roll data
x, color = make_swiss_roll(n_samples=1500)

# Apply Locally Linear Embedding
lle = LocallyLinearEmbedding(n_neighbors=10, n_components=2, method='standard')
x_lle = lle.fit_transform(x)

# Plot
plt.scatter(x_lle[:, 0], x_lle[:, 1], c=color, cmap='Spectral')
plt.title('Unrolled Swiss Roll using LLE')
plt.xlabel('LLE1')
plt.ylabel('LLE2')
plt.colorbar(label='Color mapping')
plt.show()
```
