```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Time Series Visualization

# https://www.kaggle.com/datasets/berkeleyearth/climate-change-earth-surface-temperature-data

ds = pd.read_csv('/content/drive/MyDrive/Datasets/GlobalLandTemperaturesByMajorCity.csv')
sao_paulo = ds[ds['City'] == 'São Paulo']
step = 242

plt.figure(figsize=(8, 6))
plt.plot(sao_paulo['dt'][::step], sao_paulo['AverageTemperature'][::step], color='#f5c70f')
plt.xticks(rotation=45)
plt.xlabel('Date')
plt.ylabel('Temperature (ºC)')
plt.title('Average Temperature of São Paulo over the Years')
plt.show()
```

```python
# Multi-variable Comparison

# https://www.kaggle.com/harlfoxem/housesalesprediction

ds = pd.read_csv('/content/drive/MyDrive/Datasets/kc_house_data.csv')
ds.drop(15870, axis=0, inplace=True)

plt.figure(figsize=(8, 8))
sns.scatterplot(x='bedrooms', y='price', data=ds, s=8, color='#f5c70f')
sns.regplot(x='bedrooms', y='price', data=ds, scatter=False, line_kws={'color':'black', 'linewidth':1})
plt.xlabel('Number of Bedrooms')
plt.ylabel('Price in Million of Dollars')
plt.title('House price by number of bedrooms')
plt.show()
```

```python
# Histograms for Distribution Analysis
import scipy.stats as stats

ds = pd.read_csv('/content/drive/MyDrive/Datasets/adult.data', names=np.arange(16))

plt.hist(ds[0], color='#f5c70f', density=True)

mean, std = np.mean(ds[0]), np.std(ds[0])
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mean, std)

plt.plot(x, p, 'k', linewidth=1)

plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Frequency of Age in Dataset')
plt.show()
```

```python
# Pie Chart for Category Dominance

brands = 'Apple Samsumg Huawei Others'.split()
market_share = [40, 30, 15, 15]
colors = 'lightgray darkblue red #f5c70f'.split()
explode = [0.1, 0, 0, 0]

plt.figure(figsize=(8, 8))
plt.pie(market_share, labels=brands, startangle=90, colors=colors, explode=explode, autopct='%1.1f%%')
plt.title('Pie chart of smartphone market share')
plt.show()
```

```python
# Dynamic Visualization with Sliders

from matplotlib.widgets import Slider

years = np.arange(2010, 2021)
gdp_growth = {country: np.random.rand(11) * 6 - 3 for country in ['USA', 'China' 'India', 'Brazil']}

fig, ax = plt.subplots()
plt.subplots_adjust(bottom=0.25)
l = plt.plot(years, gdp_growth['USA'], lw=2, color='#f5c70f')
plt.title('Annual GDP Growth Rate')
plt.xlabel('Year')
plt.ylabel('GDP Growth (%)')

axpos = plt.axes([0.2, 0.1, 0.65, 0.03], facecolor='lightyellow')
spos = Slider(axpos, 'Year', 2010, 2020, valinit=2010)

def update(val):
  year = int(spos.val)
  l.set_ydata(gdp_growth['USA'][year-2010])
  fig.canvas.draw_idle()

spos.on_changed(update)
plt.show()
```

```python
# Heatmaps for Correlation Analysis

ds = np.random.rand(10, 10)

plt.figure(figsize=(8, 6))
sns.heatmap(np.corrcoef(ds), annot=True, cmap='Wistia')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
# Error Bars for Data Accuracy

np.random.seed(8)
months = np.arange(1, 13)
sales = np.random.rand(12) * 1000 + 200
errors = np.random.rand(12) * 100

plt.figure(figsize=(8, 6))
plt.errorbar(months, sales, yerr=errors, fmt='o', color='#f5c70f', ecolor='black', capsize=2)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Errorbar with sales by month')
plt.show()
```

```python
# Subplots for Comparative Visualization

np.random.seed(0)
species = 'Lion Tiger Bear'.split()
weights = {s: np.random.normal(loc=300, scale=50, size=50) for s in species}

fig, axs = plt.subplots(1, 3, figsize=(15, 5), sharey=True)
for ax, specie in zip(axs, species):
  ax.hist(weights[specie], bins=15, alpha=0.7, color='#f5c70f')
  ax.set_title(f'{specie} Weights')
  ax.set_xlabel('Weight (kg)')
  ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# 3D Surface Plots for Multivariate Data

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 100) # limits and how much values between then
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

surf = ax.plot_surface(x, y, z, cmap='Wistia')
fig.colorbar(surf)
ax.set_title('3D Surface Plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```
