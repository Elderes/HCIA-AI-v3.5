```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
# Basic plot

ds = pd.read_csv(r'n8mc-b4w4.csv')
dates, cases = np.unique(ds['case_month'], return_counts=True)

plt.figure(figsize=(18, 6))
plt.plot(dates, cases, color='y')
plt.xticks(rotation=45)
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.title('COVID-19 Cases by Month and Year in United States')
plt.grid(True, alpha=0.5)
plt.show()
```

```python
# Scatter plot

# https://www.gapminder.org/data/ -> GDP per capta
# https://www.gapminder.org/data/ -> Life expectancy

life_expectancy = pd.read_csv(r'/content/drive/MyDrive/Datasets/lex.csv')
gdp_per_capta = pd.read_csv(r'/content/drive/MyDrive/Datasets/gdp_pcap.csv')

plt.figure(figsize=(12, 6))
plt.scatter(life_expectancy.iloc[23, 1:],
            gdp_per_capta.iloc[23, 1:]
            .replace({'k': '*1e3'}, regex=True)
            .map(pd.eval),
            s=2,
            c=life_expectancy.iloc[23, 1:])
plt.xlabel('Life expectancy')
plt.ylabel('GDP per capta')
plt.title('Relation between life expectancy and GDP per capta in Brazil')
plt.show()

# Try with 10 years interval
```

```python
# Bar chart

# https://www.bls.gov/oes/tables.html

ds = pd.read_excel('/content/drive/MyDrive/Datasets/all_data_M_2023.xlsx')

area, counts = np.unique(ds['OCC_TITLE'], return_counts=True)

plt.figure(figsize=(20, 6))
plt.bar(ds.OCC_TITLE, ds.H_MEAN)
plt.xticks(rotation=90, size=8)
plt.show()
```

```python
# Pie chart

# U.S. total energy statistics - https://www.eia.gov/energyexplained/us-energy-facts/data-and-statistics.php

ds = [['Petroleum', 'Natural Gas', 'Renewable energy', 'Coal', 'Nuclear eletric power'],
      [36, 33, 13, 10, 8]]

energy_source = ds[0]
share_of_total = ds[1]

plt.figure(figsize=(8, 8))
plt.pie(share_of_total, labels=energy_source, colors=plt.cm.Paired.colors, autopct='%1.1f%%')
plt.title('Pie chart of energy production sources in United States')
plt.legend(loc='lower left')
plt.show()
```

```python
# Subplots

# https://www.gapminder.org/data/ -> GDP per capta
# https://www.gapminder.org/data/ -> population trend

gdp_per_capta = pd.read_csv(r'/content/drive/MyDrive/Datasets/gdp_pcap.csv')
population = pd.read_csv(r'/content/drive/MyDrive/Datasets/pop.csv')

gdp = pd.DataFrame(gdp_per_capta.iloc[:, 1:].replace({'k':'1000'}, regex=True).applymap(pd.eval))
brazil_gdp = gdp.iloc[23, :]
eua_gdp = gdp.iloc[183, :]

population = population.fillna(0)
pop = pd.DataFrame(population.iloc[:, 1:].replace({'k':'1000', 'M':'1000000', 'B':'1000000000'}, regex=True).applymap(pd.eval))
brazil_pop = pop.iloc[23, :]
eua_pop = pop.iloc[183, :]

fig, ax = plt.subplots(1, 2, figsize=(8, 4))
ax[0].plot(np.arange(1800, 2101), brazil_gdp)
ax[0].plot(np.arange(1800, 2101), eua_gdp)
ax[1].plot(np.arange(1800, 2101), brazil_pop)
ax[1].plot(np.arange(1800, 2101), eua_pop)

ax[0].set_title('GDP by Years')
ax[0].set_xlabel('Year')
ax[0].set_ylabel('GDP')
ax[0].legend(labels=['brazil', 'eua'])

ax[1].set_title('Population by Years')
ax[1].set_xlabel('Year')
ax[1].set_ylabel('Population in billion')
ax[1].legend(labels=['brazil', 'eua'])

plt.show()

# Fill wissing values with mean?
```

```python
# Time series analysis

# https://www.kaggle.com/datasets/die9origephit/temperature-data-albany-new-york

ds = pd.read_csv('/content/drive/MyDrive/Datasets/monthly_data.csv')

plt.figure(figsize=(8, 6))
plt.plot(ds['DATE'][::10], ds['MonthlyMeanTemperature'][::10], color='yellow')
plt.xticks(rotation=45)
plt.xlabel('Date')
plt.ylabel('Average Temperature')
plt.title('Average Temperature by Date in New York')
plt.show()
```

```python
# Box plot

# https://www.kaggle.com/datasets/die9origephit/temperature-data-albany-new-york

ds = pd.read_csv('/content/drive/MyDrive/Datasets/student-mat.csv')
final_grades = [ds['G1'], ds['G2'], ds['G3']]

plt.figure(figsize=(8, 6))
plt.boxplot(final_grades, labels=['G1', 'G2', 'G3'])
plt.ylabel('Final Grade')
plt.title('Distribution of Final Grades in Mathematics')
plt.show()
```
